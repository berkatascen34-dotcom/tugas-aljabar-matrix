# Lakukan tahapan pada soal 1 dengan menggunakan pytorch

```
import torch
```

```
# buat variabel skalar x dan simpan suatu nilai skalar
x = torch.tensor(82)
```

```
# tampilkan tipe data dari x
print("Nilai x:", x)
print("Tipe data x:", x.dtype)
```

```
# buat variabel skalar y dan simpan suatu nilai skalar
y = torch.tensor(35)
```

```
# tampilkan tipe data dari y
print("\nNilai y:", y)
print("Tipe data y:", y.dtype)
```

```
# jumlahkan x + y
hasil = x + y
print("\nHasil x + y:", hasil)
```

```
# buat variabel bernama x_float dan simpan suatu nilai desimal
x_float = torch.tensor(8.2)
```

```
# jumlahkan x_float dengan y (perlu samakan tipe data)
y_float = y.to(torch.float32)
hasil_float = x_float + y_float
```

```
print("\nNilai x_float:", x_float)
print("Tipe data x_float:", x_float.dtype)
```

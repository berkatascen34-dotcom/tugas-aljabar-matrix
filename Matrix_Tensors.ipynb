**Numpy**

```python
import numpy as np
```

```python
# Buat matrix X ukuran 2x3
X = np.array([[1, 2, 3],
              [4, 5, 6]])
```

```python
print("Matrix X:")
print(X)
```

```python
# Tampilkan ukuran matrix
print("\nUkuran matrix X:", X.shape)
```

```python
# Kolom pertama
print("\nKolom pertama:")
print(X[:, 0])
```

```python
# Baris kedua
print("\nBaris kedua:")
print(X[1, :])
```

**Pytorch**

```python
import torch
```

```python
# Buat matrix X ukuran 2x3
X = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])
```

```python
print("Matrix X:")
print(X)
```

```python
# Ukuran matrix
print("\nUkuran matrix X:", X.shape)
```

```python
# Kolom pertama
print("\nKolom pertama:")
print(X[:, 0])
```

```python
# Baris kedua
print("\nBaris kedua:")
print(X[1, :])
```

**Tensorflow**

```python
import tensorflow as tf
```

```python
# Buat matrix X ukuran 2x3
X = tf.constant([[1, 2, 3],
                 [4, 5, 6]])
```

```python
print("Matrix X:")
print(X)
```

```python
# Ukuran matrix
print("\nUkuran matrix X:", X.shape)
```

```python
# Kolom pertama
print("\nKolom pertama:")
print(X[:, 0].numpy())
```

```python
# Baris kedua
print("\nBaris kedua:")
print(X[1, :].numpy())
```

```
import numpy as np
```

```
# 1. Buat vector x ukuran 3x1
x = np.array([30, 5, 10])
x = x.reshape(3, 1)
```

```
print("Vector x:")
print(x)
```

```
# Hitung nilai L2 norm dari vector x
norm_x = np.linalg.norm(x)
print("\nL2 norm dari vector x:", norm_x)
```

-----------------------------

```
# 2. Buat vector v dengan elemen (3, 4)
v = np.array([3, 4])
```

```
print("\nVector v:")
print(v)
```

```
# Hitung nilai L2 norm dari vector v
norm_v = np.linalg.norm(v)
print("\nL2 norm dari vector v:", norm_v)
```

```
# Tampilkan unit vector dari vector v
unit_v = v / norm_v
print("\nUnit vector dari v:")
print(unit_v)
```

```
# Hitung panjang unit vector dari vector v
length_unit_v = np.linalg.norm(unit_v)
print("\nPanjang unit vector v:", length_unit_v)
```
